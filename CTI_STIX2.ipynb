```python
try:
    from stix2 import AttackPattern
except Exception as e:
    print (e)
else:
    attackpattern=AttackPattern(name = "Spear Phishing",
                    
                    external_references=[
                        {
                            "url":"https://capec.mitre.org/data/definitions/98.html",
                            "source_name":"CAPEC",
                        },
                    ],
                    kill_chain_phases=[
                        {
                            
                        "kill_chain_name":"mand-attack-lifecycle-model",
                        "phase_name":"initial-compromise",
                        
                        }
                    ]
                )
    print(attackpattern.serialize(pretty=True))
```

```python
from stix2 import ThreatActor

ta=ThreatActor(name = "Ron Aldo",
               aliases=["Uncatchable"],
               threat_actor_types=["spy","criminal"],
               sophistication = "expert",
               resource_level = "individual",
               goals = ["Steal private files"],
               first_seen = "2015-05-07T14:22:14.760Z",
               description="Uses spear phishing attacks to deliver remote access malware to its targets",
              )
print(ta.serialize(pretty=True))
```

```python
try:
    from stix2 import Relationship
    relationship = Relationship(ta,"uses",malware)
    relationship2 = Relationship(ta,"uses",attackpattern) 
except Exception as e:
    print(f'ERROR: {e}')
else:
    print(relationship.serialize(pretty=True))
    print(relationship2.serialize(pretty=True))
```

```python
from stix2 import Bundle

try:
    bundle = Bundle(malware,ta,attackpattern,relationship,relationship2,
                    allow_custom=True,
                    )
    
except Exception as e:
   raise e
    
else:
    print(bundle.serialize(pretty=True))
```
